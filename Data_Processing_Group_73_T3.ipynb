```python
import pandas as pd
import networkx as nx
import json

# Load dataset
df = pd.read_csv('Data/students_data.csv')

# Build graph
G = nx.from_pandas_edgelist(
    df, 
    source='source_name', 
    target='target_name', 
    edge_attr=['weight', 'type', 'layer', 'is_directed'],
    create_using=nx.MultiDiGraph() 
)

# Compute Network Metrics (M2)
# Calculate Betweenness Centrality to reflect node importance
centrality = nx.betweenness_centrality(G)
nx.set_node_attributes(G, centrality, 'betweenness_centrality')

# Export to JSON for Cytoscape.js
cytoscape_data = nx.cytoscape_data(G)
with open('network_data.json', 'w') as f:
    json.dump(cytoscape_data, f)
```
